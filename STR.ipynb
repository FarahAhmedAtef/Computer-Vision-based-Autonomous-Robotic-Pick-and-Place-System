## Mars Tests

```python
import easyocr
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image_path = 'mars4.jpg'
image = cv2.imread(image_path)

reader = easyocr.Reader(['en'])

results = reader.readtext(image)

for (bbox, text, confidence) in results:
    box = [tuple(map(int, point)) for point in bbox]

    cv2.polylines(image, [np.array(box)], isClosed=True, color=(0, 255, 0), thickness=2)

    x, y = box[3]
    text_position = (x, y + 30)
    cv2.putText(image, text, text_position, cv2.FONT_HERSHEY_SIMPLEX, 1.2, (255, 0, 0), 3)

cv2_imshow(image)
```

# Mars Tests

```python
import cv2
from paddleocr import PaddleOCR
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread('/content/mars4.jpg')

ocr = PaddleOCR(use_angle_cls=True, lang='en')

result = ocr.ocr(img, cls=True)

if result:
    for line in result[0]:
        text = line[1][0]
        coordinates = line[0]

        pts = np.array(coordinates, dtype=np.int32)

        img = cv2.polylines(img, [pts], isClosed=True, color=(0, 255, 0), thickness=2)

        x, y = pts[0][0], pts[0][1]
        cv2.putText(img, text, (x, y + 30), cv2.FONT_HERSHEY_SIMPLEX,
                    1, (255, 0, 0), 2, cv2.LINE_AA)

cv2_imshow(img)
```
